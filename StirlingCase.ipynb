```python
import pandas as pd
```

```python
#Reading the excel file in Pandas
df_angleBar = pd.read_excel('./Resources/AngleBar.xlsx')
df_angleBar
```

### For loop with an if statement to generate a new Index based on column name 'BA304', 'Thickness' and 'Size'.
### Then creating a new spreadsheet with the new Index and category "Angle_Bar".

```python
BA304_index = []

for index, row in df_angleBar.iterrows():
    if row['BA304'] and row['Thickness'] and row['Size(mm)']:
        BA304_index.append(f"BA304_{row['Thickness']}_{row['Size(mm)']}")
    else:
        BA304_index.append(None)

df_angleBar['NewIndex'] = BA304_index
df_304 = df_angleBar[['NewIndex','Thickness','Size(mm)','Weight(kg/m)']]
df_304 = df_304.dropna().set_index('NewIndex')#Dropping null values and inserting the new index.
df_304['Category'] = 'Angle_Bar' #Adding the category "Angle_Bar to the table".
df_304.to_excel('./Resources/angleBar304Index.xlsx')
```

```python
#This is the result of the new 304 index Angle_Bar table.
df_304
```
